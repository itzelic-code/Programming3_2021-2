```python
movies2 = pd.read_csv('data/movie.csv', nrows = 1000,
                     dtype = {
                         'num_critic_for_reviews' : np.float16,
                         'duration' : np.float16,
                         'director_facebook_likes' : np.float16,
                         'actor_3_facebook_likes' : np.float16,
                         'actor_1_facebook_likes' : np.float16,
                         'gross' : np.float16,
                         'color' : 'category',
                         'genres' : 'category',
                         'language' : 'category',
                         'country' : 'category',
                         'content_rating' : 'category'
                     })

movies2.info()
```
